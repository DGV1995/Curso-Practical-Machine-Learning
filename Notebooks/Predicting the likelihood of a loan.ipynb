```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Preparing the dataset

```python
data = requests.get("https://sds-platform-private.s3-us-east-2.amazonaws.com/uploads/P39-Financial-Data.csv")
```

```python
lines = data.text.split("\n")
```

```python
lines = [line.replace("\r", "") for line in lines]
lines
```

```python
columns = lines[0].split(",")
columns
```

```python
data = [lines[i].split(",") for i in range(1, len(lines))]
data
```

```python
df = pd.DataFrame(data=data, columns=columns)
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.dropna(axis = 0, inplace=True)
```

```python
df.shape
```

```python
df.isna().any() # Dataset in cleaned now
```

```python
df["pay_schedule"].unique()
```

```python
df.dtypes
```

```python
float_columns = ["risk_score_2","risk_score_3","risk_score_4","risk_score_5","ext_quality_score",
                 "ext_quality_score_2"]
int_columns = [col for col in df.columns if col not in float_columns and col not in ["pay_schedule"]]
```

```python
df[int_columns].head()
```

```python
df[float_columns].head()
```

```python
df[int_columns] = df[int_columns].astype("int32")
df[float_columns] = df[float_columns].astype("float32")
```

```python
df.describe()
```

## Representating the data

```python
dataset2 = df.drop(columns=["entry_id", "pay_schedule", "e_signed"])
```

```python
dataset2.shape
```

```python
plt.figure(figsize=(20,15))
plt.suptitle("Histogram of numerical columns", fontsize=15)

i = 0

for col in dataset2.columns:
    plt.subplot(6, 3, i+1)
    plt.title(col)
    values = np.size(dataset2[col].unique())
    if values > 100:
        values = 100
    plt.hist(dataset2[col], bins=values)
    plt.subplots_adjust(hspace=0.6)
    i += 1
```
