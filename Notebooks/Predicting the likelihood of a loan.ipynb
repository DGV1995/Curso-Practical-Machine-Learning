## Feature engineering

```python
import random
random.seed(100)
```

```python
df.drop(["months_employed"], axis=1, inplace=True)
```

```python
df["personal_account_months"] = df.personal_account_m + df.personal_account_y*12
```

```python
df[["personal_account_m", "personal_account_y", "personal_account_months"]].head()
```

```python
df.drop(columns=["personal_account_m", "personal_account_y"], inplace=True)
```

## Data preprocessing

```python
df = pd.get_dummies(df)
```

```python
df.columns
```

```python
df.drop(columns=["pay_schedule_semi-monthly"], inplace=True) 
# We make the variables independent from each other again
```

```python
# Removing extra columns
response = df.e_signed
users = df.entry_id
df = df.drop(["entry_id", "e_signed"], axis=1)
```

```python
# Training and testing datasets

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df, response, test_size=0.2, random_state=0)
```

```python
# Feature scaling

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled = pd.DataFrame(data=scaler.fit_transform(X_train),
                              columns=X_train.columns.values, 
                              index=X_train.index.values)
X_test_scaled = pd.DataFrame(data=scaler.fit_transform(X_test), 
                             columns=X_test.columns.values, 
                             index=X_test.index.values)
```

```python
X_train_scaled.head()
```

## Model building

### Logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression(random_state=0, penalty="l1")
#Penalize if some variable has a very hight coefficient
```

```python
classifier.fit(X_train_scaled, y_train)
```

```python
y_pred = classifier.predict(X_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

accuracy, f1, precision, recall
```

```python
results = pd.DataFrame(data = [["Linear Regression", accuracy, precision, f1, recall]],
               columns = ["Model", "Accuracy", "Precision", "F1", "Recall"])
results
```

### SVM (linear)

```python
from sklearn.svm import SVC

classifier = SVC(random_state=0, kernel="linear")
classifier.fit(X_train_scaled, y_train)
y_pred = classifier.predict(X_test_scaled)

accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

svm_results = pd.DataFrame(data = [["SVM (linear)", accuracy, precision, f1, recall]],
               columns = ["Model", "Accuracy", "Precision", "F1", "Recall"])

results = results.append(svm_results, ignore_index=True)
results
```

### SVM (rbf)

```python
classifier = SVC(random_state=0, kernel="rbf")
classifier.fit(X_train_scaled, y_train)
y_pred = classifier.predict(X_test_scaled)

accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

svm_results = pd.DataFrame(data = [["SVM (rbf)", accuracy, precision, f1, recall]],
               columns = ["Model", "Accuracy", "Precision", "F1", "Recall"])

results = results.append(svm_results, ignore_index=True)
results
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(random_state=0, n_estimators=100) # number of trees = 100
classifier.fit(X_train_scaled, y_train)
y_pred = classifier.predict(X_test_scaled)

accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

svm_results = pd.DataFrame(data = [["Random Forest", accuracy, precision, f1, recall]],
               columns = ["Model", "Accuracy", "Precision", "F1", "Recall"])

results = results.append(svm_results, ignore_index=True)
results
```

## K-Fold Cross Validation

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(classifier, X=X_train_scaled, y=y_train, cv=10)

print("Random Forest accuracy: {:.3f} (+- {:.3f})".format(accuracies.mean(), accuracies.std()*2))
```
