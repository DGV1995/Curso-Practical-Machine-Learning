# STEP #3: TRAINING THE MODEL

```python
X_train = training[:, 1:]/255
y_train = training[:, 0]
```

```python
X_test = testing[:, 1:]/255
y_test = testing[:, 0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_validation, y_train, y_validation = train_test_split(X_train, y_train, test_size=0.2, random_state=12345)
```

```python
X_train = X_train.reshape(X_train.shape[0], *(28,28,1))
X_test = X_test.reshape(X_test.shape[0], *(28,28,1))
X_validation = X_validation.reshape(X_validation.shape[0], *(28,28,1))
```

```python
X_train.shape
```

```python
X_validation.shape
```

```python
import keras 
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from keras.optimizers import Adam
from keras.callbacks import TensorBoard
```

```python
cnn_model = Sequential()
```

```python
# Convolution network
cnn_model.add(Conv2D(32, 3, 3, input_shape=(28,28,1), activation="relu")) # 32 kernels, size 3x3
```

```python
# Pooling layer
cnn_model.add(MaxPooling2D(pool_size=(2,2))) # MaxPooling function
```

```python
# Flatten our model
cnn_model.add(Flatten())
```

```python
# Apply our dense function
cnn_model.add(Dense(output_dim=32, activation="relu"))
```

```python
cnn_model.add(Dense(output_dim=10, activation="sigmoid"))
```

```python
# Adam Optimizer
# loss = sparse_categorical_crossentropy because we are classificating into 10 categories
cnn_model.compile(loss="sparse_categorical_crossentropy", optimizer=Adam(lr=0.001), metrics=["accuracy"])
```

```python
epochs = 50
cnn_model.fit(X_train, 
              y_train, 
              nb_epoch=epochs, 
              batch_size=512, verbose=1, 
              validation_data=(X_validation, y_validation))
```
